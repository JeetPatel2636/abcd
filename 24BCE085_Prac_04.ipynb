```python
import pandas as pd
import time

mat = pd.read_csv("6citytsp.csv",header=None).values

print(mat)
print(mat.dtype)
```

```python
#Brute Force

from itertools import permutations
import numpy as np

citynames = list(range(mat.shape[0]))
per = permutations(citynames)
print(type(per))
per = list(per)
print(type(per))
print(per)
print()

st = time.process_time()
print(st)
besttourlength = np.inf

for tour in per:
    tourlength = 0
    for i in range(len(tour)-1):
        tourlength += mat[tour[i],tour[i+1]]
    tourlength += mat[tour[i+1],tour[0]]
    
    if tourlength < besttourlength:
        besttourlength = tourlength
        besttour = tour

et = time.process_time()
time_taken_ms = (et-st)*1000
print("Best Tour:", besttour)
print("Best Tour's Length:", besttourlength)
print("Time Taken: ", time_taken_ms)
```

```python
#TSP using Nearest Neighbour Heuristic, not always optimal
import pandas as pd
import numpy as np
import time

mat = pd.read_csv("11citytsp.csv",header=None).values.astype(float)

startcity = int(input("Enter Starting City from 0 to "+str(mat.shape[0]-1) + ":"))
st = time.process_time()
tourlength = 0
besttour = [startcity]
mat[mat==0] = np.inf
print(mat)

matorigcopy =  mat.copy()

for i in range(mat.shape[0]-1):
    if i==0:
        tourlength += min(mat[startcity,:])
        nextbestind = int(np.argmin(mat[startcity,:]))
        besttour.append(nextbestind)
        mat[:,nextbestind] = np.inf
        mat[:,startcity] = np.inf
    else:
        tourlength += min(mat[nextbestind,:])
        nextbestind = int(np.argmin(mat[nextbestind,:]))
        mat[:,nextbestind]=np.inf
        besttour.append(nextbestind)

tourlength += matorigcopy[nextbestind,startcity]
et = time.process_time()
time_taken_ms = (et - st)*1000
print("Best Tour:", besttour)
print("Best Tour's Length:", tourlength)
print("Time Taken: ",time_taken_ms)
```
